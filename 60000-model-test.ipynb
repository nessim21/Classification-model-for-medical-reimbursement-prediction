```python
import pandas as pd
import re

# Function to split medicine name and weight
def split_medicine_weight(medicine):
    # Use regex to match the medicine name and the weight
    if isinstance(medicine, str):  # Check if the input is a string
        match = re.match(r"([^\d]+)(.*)", medicine)
        if match:
            medicament = match.group(1).strip()
            grammage = match.group(2).strip()
            return medicament, grammage
    return None, None  # Handle non-string values

# Convert 'designation' column to string to handle non-string values
data['designation'] = data['designation'].astype(str)

# Apply the function to the 'designation' column
data[['medicament', 'grammage']] = data['designation'].apply(lambda x: pd.Series(split_medicine_weight(x)))

# Display the DataFrame to verify
print(data.head())
```

```python
# Supprimer les lignes où les colonnes 'date_dossier', 'date_pharmacie' ou 'designation' contiennent des valeurs nulles
data= data.dropna(subset=['duree', 'quantite','date_dossier', 'date_pharmacie', 'designation','libelle','grammage','matricule_meg','task_name_label'])
```

```python

data2=pd.read_excel("/kaggle/input/classes-medicaments/liste_amm_reduit.xlsx")

# Supprimer les lignes dupliquées
data2= data2.drop_duplicates()
# Supprimer les lignes vides
data2 = data2.dropna(how='all')
# Réinitialiser l'index
data2 = data2.reset_index(drop=True)
data2 = data2.rename(columns={'Nom': 'medicament'})
data2.head()
```

```python
import pandas as pd

# Define the mapping of values to replace
replacement_map = {
    '4488': 'Remboursable',
    '4484': 'Non remboursable',
    '4487': 'Non remboursable APCI'
}

# Replace the values in the 'libelle.1' column
merged_df['libelle.1'] = merged_df['libelle.1'].replace(replacement_map)

# Define the mapping of values to replace
replacement_map = {
    'Remboursable': 0,
    'Non remboursable': 1,
    'Non remboursable APCI': 2
}

# Replace the values in the 'libelle.1' column
merged_df['libelle.1'] = merged_df['libelle.1'].replace(replacement_map)
# Display the updated DataFrame
print(merged_df['libelle.1'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
```

```python
df=merged_df[['id_dossier','id_tiers_adh','matricule_meg','id_tiers_bnef','libelle','date_dossier','medicament','DCI','Classe','Sous Classe','quantite','duree','code_avis','libelle.1']]
df.shape
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Liste des colonnes à encoder
columns_to_encode = ['date_dossier','libelle','medicament','DCI','Classe','Sous Classe','quantite','duree']  # Remplacez par les noms de vos colonnes

d = dict()
for column in columns_to_encode:
    encoder = LabelEncoder()
    df.loc[:, column] = encoder.fit_transform(df[column])
    d[column] = encoder
```

```python
for c in df.columns:
    df[c]=df[c].map(lambda x:float(x))
```

```python
df
```

```python
# Préparer les caractéristiques et la cible
features = df[['libelle','medicament','DCI','Classe','Sous Classe','quantite','duree',]]
target = df['libelle.1']
```

```python
 df['libelle.1']
```

```python
# Fractionner les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(features, target,test_size=0.2,random_state=42)
```

```python
# Créer et entraîner le modèle de classification
model=XGBClassifier(random_state=42)
# model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Faire des prédictions
y_pred = model.predict(X_test)
```

```python

# Évaluer le modèle
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, features, target, cv=5
                        )
print(f'Validation Cross Scores: {scores}')
print(f'Mean Cross Validation Score: {scores.mean()}')
```
